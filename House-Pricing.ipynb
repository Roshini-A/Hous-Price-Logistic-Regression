# Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing the dataset

```python
dataset=pd.read_csv("C:\\Users\\LENOVO\\Downloads\\house_price.csv")
```

```python
dataset.head()
```

# Dummy Variables and Standardisation

```python
x=pd.get_dummies(dataset["Location"],drop_first=True)
x
```

```python
dataset=pd.concat([dataset,x],axis=1,)
```

```python
dataset
```

```python
dataset.drop("Location",axis=1,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
dataset["Sq.ft"]=scaler.fit_transform(dataset[["Sq.ft"]])
```

```python
x=dataset.loc[:,['BHK','Furnishing','Sq.ft','Old(years)','Floor','Whitefield']]
y=dataset.loc[:,['Price']]
```

```python
x
```

```python
y
```

# Splitting dataset into the Training set and Test set

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
X_train 
```

```python
Y_train
```

# Training Logistic Regression Model

```python
#fitting linear regression to the training set
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(X_train,Y_train)
```

```python
X_train
```

```python
Y_train
```

# Predicting new test result

```python
y_pred=reg.predict(X_test)
y_pred
```

```python
Y_test
```

```python
#accuracy test
from sklearn.metrics import r2_score
r2_score(Y_test,y_pred)
```
